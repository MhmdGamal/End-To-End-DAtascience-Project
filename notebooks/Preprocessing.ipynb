```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pickle
```

```python
df=pd.read_csv('../data/processed/standard_correlated_df.csv')
df.head()
```

```python
df.info()
```

```python
numerical_features = [col for col in df.columns if df[col].dtype != 'object']

numerical_features.remove('life_expectancy')

numerical_features
```

```python
category_features = [col for col in df.columns if df[col].dtype == 'object']
category_features
```

```python
df_c=df.copy()
```

```python
# Preprocessing for numerical data
numerical_transformer = StandardScaler()

# Preprocessing for categorical data
categorical_transformer = OneHotEncoder(drop='first')

# Bundle preprocessing for numerical and categorical data
transformer = make_column_transformer(
    (numerical_transformer, numerical_features),
    (categorical_transformer, category_features),remainder="passthrough",sparse_threshold=0)
```

```python
transformed_df=transformer.fit_transform(df_c[numerical_features+category_features])
transformed_df
```

```python
len(transformed_df[0])
```

```python
transformed_df=pd.DataFrame(transformed_df,columns=numerical_features+category_features)
transformed_df
```

```python

transformed_df['life_expectancy']=df['life_expectancy']
transformed_df
```

```python
transformed_df.to_csv('../data/processed/standard_transformed_df.csv',index=False)
```
